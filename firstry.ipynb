```python
import ee
necoast = ee.Geometry.Polygon(
        [[[-74.02587890625, 41.06231485732392],
          [-73.01513671875, 40.429747919609724],
          [-69.521484375, 41.16164344037809],
          [-70.02685546875, 42.552619913667606],
          [-70.29052734375, 42.955965122403214],
          [-69.27978515625, 43.56401874374831],
          [-67.412109375, 44.30767946955783],
          [-66.73095703125, 44.55871811691956],
          [-67.34619140625, 45.057559971566114],
          [-69.98291015625, 44.59002212257727],
          [-71.3671875, 43.579938131822814],
          [-71.4990234375, 42.341843351631155],
          [-72.685546875, 41.57389377874231]]])

imageVisParam = {"opacity":1,"bands":["B4","B3","B2"],"min":-560.1588934829267,"max":2614.3187289464204,"gamma":1}
```

```python

#Time window
startyear = 2013
endyear = 2020;
startmonth = 1
endmonth = 12

imageCollection = ee.ImageCollection("COPERNICUS/S2_SR")\
  .filterBounds(necoast)\
  .filter(ee.Filter.lt('HIGH_PROBA_CLOUDS_PERCENTAGE',50))\
  .filter(ee.Filter.calendarRange(startyear,endyear,'year'))\
  .filter(ee.Filter.calendarRange(startmonth,endmonth,'month'))\
  .sort('system:time_start')\
  .set('SENSOR_ID', 'OLI_TIRS')

#old s2 cloudmask
def cloudmask(img):
                  mask = img.select(['SCL']).eq(6)
                  notcloud=img.select(['QA60']).eq(0)
                  img=img.updateMask(mask)
                  img=img.updateMask(notcloud)
                  return img

imageCollection = imageCollection.map(cloudmask)

Map.addLayer(imageCollection.median(),imageVisParam,"result1")

```
